```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Superstore sales.csv')
```

```python
df.head()
```

```python
df.dtypes
```

Converting data types

```python
df = df.convert_dtypes()

df['Order Date'] = pd.to_datetime(df['Order Date'],dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'],dayfirst=True)
```

```python
df.dtypes
```

Checking for blank spaces

```python
df.isnull().sum()
```

Filling blank spaces

```python
df['Postal Code'] = df['Postal Code'].fillna(-1)
```

```python
missing_codes = df[df['Postal Code']==-1]
```

```python
missing_codes.value_counts().sum()
```

```python
df.isnull().sum()
```

Checking for duplicates

```python
df.duplicated().sum()
```

```python
df['Customer ID'].duplicated().sum()
```

Yes,If we check for duplicates of all columns, there is no duplicates found, Then if we check for each column definitely duplicates were found,Actually it's not duplicates just customers ordered another items from same postalcode/city/same name,There is no rule to one customer only order one item.

Final output

```python
df['Sales'] = df['Sales'].round(2)
```
